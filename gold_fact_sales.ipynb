```python
query = """
SELECT
    sd.order_num,
    pr.product_key,
    cu.customer_key,
    sd.order_date,
    sd.ship_date,
    sd.due_date,
    sd.sales,
    sd.quantity,
    sd.price
FROM silver.crm_salesdetails sd
LEFT JOIN gold.product_dim pr
    ON sd.product_key = pr.product_number
LEFT JOIN gold.dim_customers cu
    ON sd.customer_id = cu.customer_id;
"""
df = spark.sql(query)
df.display()
```

```python
df.write.mode("overwrite").format("delta").saveAsTable('workspace.gold.fact_sales')
```
