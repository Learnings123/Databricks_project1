```python
from pyspark.sql.functions import *
from pyspark.sql.types import*
```

```python
df=spark.table('workspace.bronze.dwh_px_cat_g1v2')
```

## **Trimming Fields**

```python

for field in df.schema.fields:
  if isinstance(field.dataType,StringType):
      df=df.withColumn(field.name,trim(col(field.name)))
df.display()
```

## **Transform Maintenance field**

```python
mapping={'No':'False','Yes':'True'}
df.replace(mapping,subset=['MAINTENANCE']).display()
```

## **Renaming Fields**

```python
Rename={'ID':'category_id','CAT':'category','SUBCAT':'sub_category','MAINTENANCE':'maintenance'}

for old,new in Rename.items():
    df=df.withColumnRenamed(old,new)
df.display()
```

## **Write to Silver**

```python
df.write.mode('overwrite').format('delta').saveAsTable('workspace.silver.erp_product_category')
```
