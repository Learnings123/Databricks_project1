Creating a config file that has all the sources path and destination table name.

```python
Ingestion_config=[
    {
        "source":"crm",
        "path":"/Volumes/workspace/dwh/dwh/source_crm/source_crm/cust_info.csv",
        "table":"dwh_cust_info"
    },
    {
        "source":"crm",
        "path":"/Volumes/workspace/dwh/dwh/source_crm/source_crm/prd_info.csv",
        "table":"dwh_prd_info"
    },
    {
        "source":"crm",
        "path":"/Volumes/workspace/dwh/dwh/source_crm/source_crm/sales_details.csv",
        "table":"dwh_sales_details"
    },
    {
        "source":"erp",
        "path":"/Volumes/workspace/dwh/dwh/source_erp/source_erp/CUST_AZ12.csv",
        "table":"dwh_CUST_AZ12"
    },
    {
        "source":"erp",
        "path":"/Volumes/workspace/dwh/dwh/source_erp/source_erp/LOC_A101.csv",
        "table":"dwh_LOC_A101"
    },
    {
        "source":"erp",
        "path":"/Volumes/workspace/dwh/dwh/source_erp/source_erp/PX_CAT_G1V2.csv", 
        "table":"dwh_PX_CAT_G1V2"
     }
]
```

```python
for item in Ingestion_config:
    print(f'ingesting from source {item['source']} to table {item['table']}')
    df=spark.read.format("csv").option("header",True).option("inferSchema",True).load(item['path'])
    df.write.mode("overwrite").format("delta").saveAsTable(f"workspace.bronze."+item['table'])
    print(f"{item['table']} Ingestion completed")
```
