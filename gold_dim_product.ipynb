```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
query = """
SELECT
    ROW_NUMBER() OVER (ORDER BY pn.start_date, pn.product_number) AS product_key, -- Surrogate key
    pn.product_id,
    pn.product_number,
    pn.product_name,
    pn.category_id,
    pc.category,
    pc.sub_category,
    pc.maintenance,
    pn.product_line,
    pn.start_date
FROM silver.crm_product pn
LEFT JOIN silver.erp_product_category pc
    ON pn.category_id = pc.category_id
--WHERE pn.end_date IS NULL; -- Filter out all historical data
"""
```

```python
df=spark.sql(query)
df.display()
```

```python
df.write.mode("overwrite").format('delta').saveAsTable("gold.product_dim")
```
